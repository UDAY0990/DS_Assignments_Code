```python
tweets=pd.read_csv('Elon_musk.csv',encoding='Latin-1')
tweets
```

```python
tweets.drop(['Unnamed: 0'],inplace=True,axis=1)
tweets
```

```python
# Text Processing
tweets=[Text.strip() for Text in tweets.Text] # remove both the leading and the trailing characters
tweets=[Text for Text in tweets if Text] # removes empty strings, because they are considered in Python as False
tweets[0:5]
```

```python
# Joining the list into one string/text
tweets_text=' '.join(tweets)
tweets_text
```

```python
# remove Twitter username handles from a given twitter text. (Removes @usernames)
from nltk.tokenize import TweetTokenizer
tkn = TweetTokenizer(strip_handles=True)
tweets_tokens=tkn.tokenize(tweets_text)
print(tweets_tokens)
```

```python
# Again Joining the list into one string/text
tweets_tokens_text=' '.join(tweets_tokens)
tweets_tokens_text
```

```python
# Remove Punctuations 
non_punc_text =tweets_tokens_text.translate(str.maketrans('','',string.punctuation))
non_punc_text
```

```python
# remove https or url within text
import re
non_url_text=re.sub(r'http\S+', '', non_punc_text)
non_url_text
```

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
```

```python
from nltk.tokenize import word_tokenize
text_tokens=word_tokenize(non_url_text)
print(text_tokens)
```

```python
# Tokens count
len(text_tokens)
```

```python
# Remove Stopwords
from nltk.corpus import stopwords
my_stop_words=stopwords.words('english')

sw_list = ['\x92','rt','ye','yeah','haha','Yes','U0001F923','I','the','an','is']
my_stop_words.extend(sw_list)

no_stop_tokens=[word for word in text_tokens if not word in my_stop_words]
print(no_stop_tokens)
```

```python
# Normalize the data
lower_words=[Text.lower() for Text in no_stop_tokens]
print(lower_words[100:200])
```

```python
# Stemming (If Required)
from nltk.stem import PorterStemmer
ps=PorterStemmer()
stemmed_tokens=[ps.stem(word) for word in lower_words]
print(stemmed_tokens[100:200])
```

```python
# Lemmatization
nlp=spacy.load('en_core_web_sm')
doc=nlp(' '.join(lower_words))
print(doc)
```

```python
lemmas=[token.lemma_ for token in doc]
print(lemmas)
```

```python
clean_tweets=' '.join(lemmas)
clean_tweets
```

WORD CLOUD

```python
# Define a function to plot word cloud
def plot_cloud(wordcloud):
    plt.figure(figsize=(15,10))
    plt.imshow(wordcloud)
    plt.axis('off')
    
# Generate Word Cloud

STOPWORDS.add('pron')
STOPWORDS.add('rt')
STOPWORDS.add('yeah')
wordcloud=WordCloud(width=3000,height=2000,background_color='white',max_words=25,
                   colormap='Set1',stopwords=STOPWORDS).generate(clean_tweets)
plot_cloud(wordcloud)
```

```python
# Parts Of Speech (POS) Tagging
nlp=spacy.load('en_core_web_sm')

one_block=clean_tweets
doc_block=nlp(one_block)
spacy.displacy.render(doc_block,style='ent',jupyter=True)
```

```python
for token in doc_block[100:150]:
    print(token,token.pos_)    
```

```python
# Filtering the nouns and verbs only
nouns_verbs=[token.text for token in doc_block if token.pos_ in ('NOUN','VERB')]
print(nouns_verbs[100:150])
```

```python
# Counting the noun & verb tokens
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()

X=cv.fit_transform(nouns_verbs)
sum_words=X.sum(axis=0)

words_freq=[(word,sum_words[0,idx]) for word,idx in cv.vocabulary_.items()]
words_freq=sorted(words_freq, key=lambda x: x[1], reverse=True)

wd_df=pd.DataFrame(words_freq)
wd_df.columns=['word','count']
wd_df[0:25] # viewing top ten results
```

```python
# Visualizing results (Barchart for top 15 nouns + verbs)
wd_df[0:15].plot.bar(x='word',figsize=(10,8),title='Top 15 nouns and verbs');
```

Emotion Mining -  Sentiment Analysis

```python
from nltk import tokenize
sentences=tokenize.sent_tokenize(' '.join(tweets))
sentences
```

```python
sentiment = pd.DataFrame(sentences,columns=['sentence'])
sentiment
```

```python
# Emotion Lexicon - Affin
from google.colab import files
uplloaded = files.upload()
afinn =pd.read_csv('Afinn.csv',sep=',',encoding='Latin-1')
afinn
```

```python
afinity_scores = afinn.set_index('word')['value'].to_dict()
afinity_scores
```

```python
# Custom function: score each word in a sentence in lemmatised form, but calculate the score for the whole original sentence
nlp=spacy.load('en_core_web_sm')
sentiment_lexicon=afinity_scores

def calculate_sentiment(text:str=None):
    sent_score=0
    if text:
        sentence=nlp(text)
        for word in sentence:
            sent_score+=sentiment_lexicon.get(word.lemma_,0)
    return sent_score
```

```python
# manual testing
calculate_sentiment(text='great')
```

```python
# Calculating sentiment value for each sentence
sentiment['sentiment_value']=sentiment['sentence'].apply(calculate_sentiment)
sentiment['sentiment_value']
```

```python
# how many words are there in a sentence?
sentiment['word_count']=sentiment['sentence'].str.split().apply(len)
sentiment['word_count']
```

```python
sentiment.sort_values(by='sentiment_value')
```

```python
# Sentiment score of the whole review
sentiment['sentiment_value'].describe()
```

```python
# positive sentiment score of the whole review
sentiment[sentiment['sentiment_value']>0]
```

```python
# negative sentiment score of the whole review
sentiment[sentiment['sentiment_value']<=0]
```

```python
sentiment['index']=range(0,len(sentiment))
sentiment
```

```python
# Correlation analysis
sentiment.plot.scatter(x='word_count',y='sentiment_value',figsize=(6,6),title='Sentence sentiment value to sentence word count')
```
