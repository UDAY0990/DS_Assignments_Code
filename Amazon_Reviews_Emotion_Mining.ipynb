```python
data=pd.read_csv('reviews.csv')
data
```

```python
data=data.rename(columns={'Review Content':'Review_Content'})
data
```

```python
# Converting to lower case
data.Review_Content=data.Review_Content.apply(lambda x: x.lower())
data
```

```python
# Splitting the words from Sentence.
data.Review_Content=data.Review_Content.apply(lambda x: x.split())
data
```

```python
# Removing Stopwords
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
nltk.download('stopwords')
```

```python
stopwords.words("english")
```

```python
data.Review_Content=data.Review_Content.apply(lambda x: [word for word in x if word not in set(stopwords.words('english'))])
data
```

```python
# Stemming
ps=PorterStemmer()
data.Review_Content=data.Review_Content.apply(lambda x: [ps.stem(word) for word in x])
data
```

```python
# Rejoining
data.Review_Content=data.Review_Content.apply(lambda x: " ".join(x))
data
```

```python
from textblob import TextBlob
```

```python
data['Sentiment']=data.Review_Content.apply(lambda x: TextBlob(x).sentiment[0])
data
```

```python
def sentiment_category(Sentiment):
    if (Sentiment > 0.4):
        return 'Positive'
    elif (Sentiment < -0.5):
        return 'Negative'
    else:
        return 'Neutral'
```

```python
data['Category']=data['Sentiment'].apply(lambda x:sentiment_category(x))
data
```

```python
data['Category'].value_counts()
```
